# Data Cleaning - Iris Dataset

```
import pandas as pd
df = pd.read_csv('iris.csv')
print(df.info())
```

```
# Missing Values
print(df.isnull().sum())
```

```
# Duplicate Records
print('Duplicates:', df.duplicated().sum())
```

```
# Remove Duplicates
df = df.drop_duplicates()
print(df.shape)
```
